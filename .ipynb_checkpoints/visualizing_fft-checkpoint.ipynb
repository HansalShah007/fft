```python
import librosa

# Load the audio file; this automatically resamples to 22,050 Hz
audio, sample_rate = librosa.load('audio_files/River Flows in You.wav', sr=None)
print(len(audio))
print(sample_rate)
```

```python
import numpy as np

# Optional: Apply a window function (e.g., Hanning)
windowed_audio = audio * np.hanning(len(audio))
```

```python
from fft import nearest_pow_2, fft

# Make sure the length of the input data is a power of two
fft_input = windowed_audio[:nearest_pow_2(len(windowed_audio))]

# Compute the FFT
frequency_spectrum = fft(list(fft_input))
```

```python
# Compute the magnitudes of the frequency spectrum
magnitudes = np.abs(frequency_spectrum)

# Frequency resolution
frequency_resolution = sample_rate / len(frequency_spectrum)

# Generate the frequency axis
frequencies = np.arange(0, sample_rate, frequency_resolution)

# Consider only the first half of the spectrum (positive frequencies)
half_index = len(frequency_spectrum) // 2
frequencies = frequencies[:half_index]
magnitudes = magnitudes[:half_index]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(frequencies, magnitudes)
plt.title('Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

```python
# Dominant frequencies

# Find the indices of the 20 largest magnitudes (dominant frequencies)
indices = np.argsort(magnitudes)[-100:]

# Sorting indices to display them in increasing order of frequency
sorted_indices = indices[np.argsort(frequencies[indices])]

# Extracting the dominant frequencies and their magnitudes
dominant_frequencies = frequencies[sorted_indices]
dominant_magnitudes = magnitudes[sorted_indices]

# Plotting
plt.figure(figsize=(10, 6))
plt.stem(dominant_frequencies, dominant_magnitudes, 'b', markerfmt='bo', basefmt="r-")
plt.title('Top 20 Dominant Frequencies')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()

# Displaying the frequencies and their corresponding magnitudes
for freq, mag in zip(dominant_frequencies, dominant_magnitudes):
    print(f"Frequency: {freq:.2f} Hz, Magnitude: {mag:.2f}")
```
